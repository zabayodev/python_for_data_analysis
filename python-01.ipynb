`2D arrays - matrices`

$$M = \begin{pmatrix}
1 & 2 & 0 \\
0 & -1 & 0 \\
2 & 1 & 4
\end{pmatrix}
$$

````python
import numpy as np
M = np.array([[1, 2, 0], [0, -1, 0], [2, 1, 4]])
A = np.array([[2, -1, 0], [1, -1, 0], [0, 1, 4]])
print(M)
print(A)
````

````python
print(M.ndim) # to get the number of dimensions
print(M.size) # number of elements of M
print(M.shape) # number of rows and columns of M
````

Matrix operations

````python
M + A # addition (M and A must have the same dimensions)
````

````python
print(M @ A) # matrix multiplication or use:
print(np.dot(M, A))
````

````python
M - A # subtraction
````

````python
M1 = np.array([[1, 2, 3], [4, 5, 6]])
M2 = np.array([[10, 11], [20, 21], [30, 31]])
M1 @ M2
````

$M/A \equiv MA^{-1}$

Now, how do we calculate $A^{-1}$ in python? 
We use `np.linalg.inv(A)` and to verify that it's the inverse of $A$, $AA^{-1}$ should give the indentity matrix.

````python
Ainv = np.linalg.inv(A)
print(np.dot(A, Ainv).astype(int))
print((A @ Ainv).astype(int)) # means A divided by itself.
````

`if`: Flow controls

```python
if condition:
    # do something
elif condition:
    # do something
else:
    # do something
```

`for loop`

```python
for element in a_list_of_elements:
    # do something to the element
```

````python
Gender_recoded = []
for gender in data['Gender']:
    if str(gender).upper() == "F":
        Gender_recoded.append("Female")
    elif gender == "female":
        Gender_recoded.append("Female")
    elif str(gender).upper() == "M":
        Gender_recoded.append("Male")
    else:
        Gender_recoded.append(gender)
````

````python
Gender_recoded
````

````python
data['Gender_cleaned'] = Gender_recoded # adding a new column
data['Gender_cleaned'].value_counts().\
    plot(kind="pie", autopct = "%1.0f%%")
plt.show()
````

Recoding using pandas' `replace` function

````python
data['Gender_cleaned2'] = data['Gender'].\
    replace({"m":"Male", "M":"Male", 
             "female":"Female", "F":"Female"}) 

# Exercise: Add a new column named BMI = Weight/(Height)**2
data["BMI"] = data["Weight"]/(data["Height"] ** 2)
data
````
