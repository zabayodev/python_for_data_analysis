````python
Gender_recorded = []
````

` for loop `
``` Python
for element in a_list_of_elements:
    # do something

````python
Gender_recorded = []
for gender in data["Gender"]:

    if str(gender).upper() == "F":
        Gender_recorded.append("Female")

    elif gender == "female":
        Gender_recorded.append("Female")

    elif str(gender).upper() == "M":
        Gender_recorded.append("Male")
    
    else:
        Gender_recorded.append(gender)
````

````python
Gender_recorded
````

````python
import matplotlib.pyplot as plt
import seaborn as sns
data["Gender_cleaned"] = Gender_recorded
data["Gender_cleaned"].value_counts().\
    plot(kind="pie", autopct = "%1.1f%%")
plt.show()
````

Using replace for data cleaning in python

````python
Gender_weight = data.groupby("Gender_cleaned")["Weight"].mean()
Gender_weight.plot(kind="bar")
plt.title("Average Weight by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Male and Female")
plt.xticks(rotation=0)
plt.show()
````
